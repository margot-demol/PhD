# SVP SCRIPPS ok

```python
files = glob(os.path.join(download_dir, "svp_scripps", "*"))
DF = []
for f in files:
    df = (
        pd.read_csv(f)
        .rename(
            columns={
                "Platform-ID": "id",
                " Timestamp(UTC)": "time",
                " GPS-Latitude(deg)": "lat",
                " GPS-Longitude(deg)": "lon",
            }
        )
        .drop("Unnamed: 8", axis=1)
    )
    df["time"] = pd.to_datetime(df["time"])
    DF.append(df.dropna())
df = pd.concat(DF).set_index("id")
tmax = str(df.time.max()).replace("-", "").replace(":", "").replace(" ", "_")
df.to_csv(os.path.join(raw_dir, f"svp_scripps_{tmax}.csv"))
```

# eodyn

```python
files = glob(os.path.join(download_dir, "eOdyn*raw*"))
print(files)
ds = xr.open_dataset(files[0])
df = ds.to_dataframe()
df["time"] = pd.to_datetime(df["time"])
df = df.rename(
    columns={"latitude": "lat", "longitude": "lon", "buoy_label": "id"}
).dropna()
df["id"] = df["id"].astype(int)
tmax = str(df.time.max()).replace("-", "").replace(":", "").replace(" ", "_")
df[["lon", "lat", "time", "id", "hs"]].set_index("time").to_csv(
    os.path.join(raw_dir, f"L0_melodi_eodyn_{tmax}.csv")
)
```
