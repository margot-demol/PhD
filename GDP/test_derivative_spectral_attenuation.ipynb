## finite difference estimations of derivatives, spectral signature

Consider a time series $u(t)$ which may be expressed as a function of its Fourier decomposition:

$
\begin{align}
u(t) &= \sum_\omega \hat u(\omega) e^{i \omega t}
\end{align}
$

We're interested in the finite difference estimation of its first order derivative:

$
\begin{align}
\frac{d}{dt}u &= \frac{u(t+dt) - u(t-dt)}{2dt},  \\
&= \sum_\omega \hat u(\omega) e^{i \omega t} \big [Â e^{i \omega dt} - e^{-i \omega dt} \big ]/2dt, \\
&= \sum_\omega \hat u(\omega) i\omega e^{i \omega t} \times \sin (\omega dt) /\omega dt, \\
&= \sum_\omega \hat u(\omega) i\omega e^{i \omega t} \times H(\omega dt),
\end{align}
$

where $H(x) = \sin(x)/x$.
In a practical situation $0\le\omega\le\pi/dt$, hence $0\le \omega dt\le \pi$.
This will lead to an attenuation of the spectrum by a factor $H^2$.


```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.01, np.pi, 0.01)
H = np.sin(x) / x
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, H**2)
ax.set_yscale("log")
ax.set_xlabel(r"$\omega dt$")
ax.grid()
ax.set_ylim(1e-4, 1)
ax.set_title("spectral attenuation")
ax.axvline(np.pi / 2, color="k")
ax.axvline(np.pi / 4, color="k")
```
